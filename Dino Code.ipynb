## Dino code

To "run" a cell, click the triangle button to the left of the code.

Run the next cell to display the text box to enter your note. Then edit your note and run the following cell to see the result. You can copy and paste that image into an email or any place you want!

```python
from IPython.display import Image
import dinocode
import ipywidgets as widgets

text = widgets.Textarea(
    value="type your dino message here...",
    description='Message:',
    layout={'height': '200px', 'width': '400px'}
)
display(text)
```

When you run the next cell, the message in the text box will be turned into dino code.

```python
dinocode.encode(text.value)
```
